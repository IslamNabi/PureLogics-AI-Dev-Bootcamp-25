#Decision Tree Classifier Lab


#Load essential libraries

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
from sklearn.model_selection import train_test_split
```

#Import the dataset

```python
df = pd.read_csv("titanic.csv")
df
```

##Drop meaningless information

```python
df = df.drop(columns = ['PassengerId','Name','SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis = 'columns')
df
```

##Data Preprocessing

```python
import math
mean_age = df.Age.mean()
mean_age = math.floor(mean_age)
print(mean_age)
```

```python
df['Age'] = df.Age.fillna(mean_age)
df
```

##Label Encode the non numeric features

```python
LE_sex = LabelEncoder()
df['Sex'] = LE_sex.fit_transform(df['Sex'])

df
```

## Define Feature Matrix (X) and Target Vector (y)


```python
y = df['Survived']
X = df.drop('Survived', axis = 'columns')
X
```

## Train-Test Split and Model Initialization


```python
model = tree.DecisionTreeClassifier()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4)
```

##Train the model

```python
model.fit(X_train, y_train)

model.score(X_test, y_test)
```

##Predict results

```python
model.predict(X_test)
```

##Try decision tree classifer on the same target column but this time use more feature columns than before.

```python
#Code here

# Load the dataset again to start fresh
df = pd.read_csv("titanic.csv")
df.head(5)
```

```python
features_to_keep = ['Pclass', 'Sex', 'Age', 'Fare', 'SibSp', 'Parch', 'Embarked']
df = df[features_to_keep + ['Survived']] 
df.head(5)
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'])
```

```python
# Define X and y
X = df.drop('Survived', axis=1)
y = df['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model_2 = tree.DecisionTreeClassifier()
```

```python
model_2.fit(X_train, y_train)

model_2.score(X_test, y_test)
```

```python
model_2.predict(X_test)
```
