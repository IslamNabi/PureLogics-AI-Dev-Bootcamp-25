```python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
car_sales["Price"] = car_sales["Price"].str.replace("[\$\,\.]", "")
car_sales["Price"] = car_sales["Price"].str[:-2]
```

```python
import pandas as pd

# Clean the Price column by removing $ and commas, then convert to numeric
car_sales["Price"] = (
    car_sales["Price"]
    .str.replace('[\$,]', '', regex=True)  # Remove $ and commas
    .astype(float)  # Convert to float (handles decimals if any)
)

# Now calculate cumulative sum
car_sales["Total Sales"] = car_sales["Price"].cumsum()

# Add Sale Date column (using today's date as start)
from datetime import datetime
car_sales["Sale Date"] = pd.date_range(datetime.today().strftime("%d/%m/%Y"),
                          periods=len(car_sales))

# View the DataFrame
car_sales
```

More specifially we'll see how age and cholesterol combined effect the target in **patients over 50 years old**.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create DataFrame with patients over 50 years old
over_50 = heart_disease[heart_disease["age"] > 50]

# Create the plot with larger figure size
fig, ax = plt.subplots(figsize=(12, 8))

# Create a more informative scatter plot
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter",  # Better color scheme
                    alpha=0.7,  # Slight transparency
                    s=100,  # Larger points
                    edgecolor="black")  # Point borders

# Customize the plot with enhanced styling
ax.set(title="Heart Disease and Cholesterol Levels for Patients Over 50",
      xlabel="Age (years)",
      ylabel="Cholesterol (mg/dl)",
      xlim=(50, 80))  # Focus on our age range

# Improve legend
legend = ax.legend(*scatter.legend_elements(),
                 title="Heart Disease",
                 loc="upper right")
legend.get_title().set_fontsize(12)

# Add mean line with label
mean_chol = over_50["chol"].mean()
ax.axhline(mean_chol,
          linestyle="--",
          color="red",
          linewidth=1,
          label=f"Mean Cholesterol ({mean_chol:.1f} mg/dl)")

# Add grid and second legend for mean line
ax.grid(alpha=0.3)
ax.legend(loc="upper left")

# Add informative text
ax.text(52, mean_chol+10,
       f"Population Average: {mean_chol:.1f} mg/dl",
       color="red",
       fontsize=10)

# Adjust layout
plt.tight_layout()
plt.show()
```
